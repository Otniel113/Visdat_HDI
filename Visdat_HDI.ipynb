## Data Preprocessing

Hanya mengambil data nasional

```python
df_hdi = df_hdi[df_hdi['Level'] == 'National']
df_hdi.shape
```

```python
df_hdi.drop(columns=['ISO_Code', 'Level', 'GDLCODE', 'Region'], inplace=True)
df_hdi.head()
```

Mengganti index menjadi Country

```python
df_hdi.set_index('Country', inplace=True)
df_hdi.head()
```

Melakukan transpos

```python
df_hdi = df_hdi.transpose()
df_hdi.head()
```

Mengubah tipe data indeks dari int menjadi str

```python
df_hdi.index = df_hdi.index.map(str)
```

Melakukan reset index sehingga Year menjadi kolom baru dan melihat tipe data kolom Year yang baru dibuat tersebut

```python
df_hdi = df_hdi.reset_index().rename(columns={'index' : 'Year'})
df_hdi.dtypes['Year']
```

Menghilangkan 'Country' dari index

```python
df_hdi.columns.name = None
df_hdi.head()
```

Melihat info dengan banyak float64 ada 186 dan object sebanyak 1

```python
df_hdi.info()
```

Melakukan Export ke Excel

```python
df_hdi.to_excel(r'HDI_clean.xlsx', index=False, header=True)
```

## Visualisasi Bokeh

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.models.widgets import Tabs, Panel
```

### Tanpa Interaksi

Membuat Column Data Source

```python
hdi_cds = ColumnDataSource(df_hdi)
```

Mengatur Tooltips untuk Hover (Jika menggerakan mouse ke gambar, menampilkan data)

```python
tooltips = [
            ('Year','$x{0}'),
            ('HDI', '$y'),
           ]
```

Tools yang dipakai

```python
select_tools = ['pan', 'wheel_zoom', 'save', 'reset']
```

Panel 1 Berisi Visualisasi Garis/Line

```python
# Mengatur Figure
fig_garis = figure(title='Compare HDI Between 2 Nations',
                   plot_height=700, plot_width=1000,
                   x_axis_label='Tahun', y_axis_label='HDI',
                   tools=select_tools)

# Negara 1
fig_garis.line('Year', 'Japan',
              color='salmon',
              legend_label='Japan',
              line_width=4,
              source=hdi_cds)

# Negara 2
fig_garis.line('Year', 'South Korea',
              color='royalblue',
              legend_label='South Korea',
              line_width=4,
              source=hdi_cds)

# Lokasi Legend
fig_garis.legend.location = 'top_left'

# Menambahkan Hover
fig_garis.add_tools(HoverTool(tooltips=tooltips))
```

Panel 2 Berisi Visualisasi Scatter / Titik

```python
# Membuat Figure
fig_titik = figure(title='Compare HDI Between 2 Nations',
                   plot_height=700, plot_width=1000,
                   x_axis_label='Tahun', y_axis_label='HDI',
                   tools=select_tools)

# Negara 1
fig_titik.scatter('Year', 'Japan',
              color='salmon',
              legend_label='Japan',
              size=10,
              source=hdi_cds)

# Negara 2
fig_titik.scatter('Year', 'South Korea',
              color='royalblue',
              legend_label='South Korea',
              size=10,
              source=hdi_cds)

# Lokasi Legend
fig_titik.legend.location = 'top_left'

# Menambahkan Hover
fig_titik.add_tools(HoverTool(tooltips=tooltips))
```

Menggabungkan kedua visualisasi tersebut

```python
# Output to file
output_file('compare_hdi_noninteraction.html', 
            title='Compare HDI Between 2 Nations')

output_notebook()

# Membuat 2 Panel
garis_panel = Panel(child=fig_garis, title='Visualisasi Garis (Line)')
titik_panel = Panel(child=fig_titik, title='Visualisasi Titik (Scatter)')

# Menggabungkan Panel menjadi Tab
tabs = Tabs(tabs=[garis_panel, titik_panel])

show(tabs)
```
